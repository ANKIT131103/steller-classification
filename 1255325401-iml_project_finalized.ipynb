```python
import seaborn as sns
import matplotlib.pyplot as plt

def pdf_plotter(df, features, target):
    for feature in features:
        plt.figure(figsize=(8, 6))
        for class_label in df[target].unique():
            subset = df[df[target] == class_label]
            sns.kdeplot(subset[feature], label=class_label, shade=True)
        plt.title(f'Probability Density Function for {feature} by {target}')
        plt.xlabel(feature)
        plt.ylabel('Density')
        plt.legend()
        plt.show()
```

```python
def box_plotter(df, features, target):
    """
    This is funtion helps in plotting the boxplot of data.
    
    Parameters
    ----------
    `df`: dataset
    `features`: columns for analysis
    `target`: target column name
    
    Returns none.
    """
    plt.figure(figsize=(15, 30))
    for (i, feature) in zip(range(len(features)), features):
        plt.subplot(int('{}2{}'.format(len(features), i+1)))
        sns.boxplot(x=target, y=feature, data=df,
                    width=0.5, flierprops={'marker': 'x'})
        plt.title(label='Boxplot of {}'.format(feature), fontsize=10)
        plt.xlabel(xlabel=None)
        plt.ylabel(ylabel=None)
        i += 1
    plt.show()
```

#Classifier 5: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Decision Tree
param_grid = {'criterion': ['gini', 'entropy'],
              'max_depth': [None, 5, 10, 15],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 2, 4]}

# Create a Decision Tree classifier
dt_classifier = DecisionTreeClassifier(random_state=0)

# Create an instance of GridSearchCV with 3-fold cross-validation
grid_search = GridSearchCV(dt_classifier, param_grid, cv=3)

# Fit the GridSearchCV on the training data to find the best hyperparameters
grid_search.fit(x_train, y_train)

# Get the best hyperparameters and train the model with them
best_params_dt = grid_search.best_params_
print("Best Hyperparameters:", best_params_dt)

# Extract hyperparameters individually
best_criterion = best_params_dt.get('criterion', 'gini')  # Default to 'gini' if 'criterion' not found
best_max_depth = best_params_dt.get('max_depth', None)
best_min_samples_split = best_params_dt.get('min_samples_split', 2)
best_min_samples_leaf = best_params_dt.get('min_samples_leaf', 1)

best_dt_model = DecisionTreeClassifier(criterion=best_criterion,
                                       max_depth=best_max_depth,
                                       min_samples_split=best_min_samples_split,
                                       min_samples_leaf=best_min_samples_leaf,
                                       random_state=0)
best_dt_model.fit(x_train, y_train)

# Evaluate the best Decision Tree model on the test set
predicted_labels_dt = best_dt_model.predict(x_test)
accuracy_dt = accuracy_score(y_test, predicted_labels_dt)

print('Best Decision Tree Model Accuracy: %.3f' % accuracy_dt)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC  # Import SVC from sklearn.svm module
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

classifiers = {
    'knn_classifier': KNeighborsClassifier(),
    'logreg_classifier': LogisticRegression(),
    'svm_clf': SVC(),
    'dt_classifier': DecisionTreeClassifier(),
    'random_forest': RandomForestClassifier()
}

classifier_name = input('Enter the name of the Classifier: ')

if classifier_name in classifiers:
    selected_classifier = classifiers[classifier_name]
    num_features = 9  # Change this to the actual number of features

    selected_classifier.fit(x_train, y_train)
   
    def classify_with_input(classifier):
        # Gather input for each feature
        features = []
        for i in range(num_features):
            feature_value = float(input(f"Enter value for feature {i + 1}: "))
            features.append(feature_value)

        # Reshape features to be in the format expected by the classifier
        input_features = [features]

        # Perform classification using the provided classifier
        prediction = classifier.predict(input_features)

        return prediction

    result = classify_with_input(selected_classifier)
    print("The classification result is:", result)
else:
    print("Invalid classifier name. Please choose from the available classifiers.")
```
